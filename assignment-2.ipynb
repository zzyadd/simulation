```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
Demand_df = pd.DataFrame.from_dict({'Demand': [0,1, 2, 3, 4], 'P': [0.10, 0.25, 0.35, 0.21,0.09]})
Demand_df['CDF'] = Demand_df['P'].cumsum()
Demand_df['range'] = Demand_df['CDF'] * 100  
Demand_df['ran']=[1] + (Demand_df["CDF"]+0.01).tolist()[:-1]

print(Demand_df.info())
Demand_df
```

```python
Lead_time_df = pd.DataFrame.from_dict({'Lead time': [1,2,3], 'P': [0.6,0.3,0.10]})
Lead_time_df['CDF'] = Lead_time_df['P'].cumsum()
Lead_time_df['range'] = (Lead_time_df['CDF'] * 10).tolist()[:-1]+[0]
Lead_time_df['ran']=([1] + (Lead_time_df["CDF"]+0.1).tolist()[:-2]+[0])
print(Lead_time_df.info())
Lead_time_df
```

```python
Demand_RN_LIST=[24,35 ,65 ,81, 54,3, 87, 27, 73, 70, 47, 45, 48, 17,9, 42, 87, 26, 36, 40, 7, 63, 19, 88, 94]
Lead_RN_LIST= [5,0,3,4,8]

M=12
```

```python
server_DF=pd.DataFrame.from_dict({'Cycle':np.repeat(range(1,6),5),'Days':range(1,26) ,"Inv. B.":0,"RN D":Demand_RN_LIST, 'Demand': 0,"Inv.E.":0,"Short":0,"Q_Order":0,"RN L":-1,"Lead":0})
Demand_conditions = [(server_DF["RN D"] >= low) & (server_DF["RN D"] <= high) for low, high in zip(Demand_df["ran"], Demand_df["range"])]
server_DF["Demand"] = np.select(Demand_conditions, Demand_df["Demand"])
server_DF=server_DF.reset_index(drop=True)

server_DF
```

```python
x=0
for i in range(4,len(server_DF),5):
    server_DF["RN L"][i]=Lead_RN_LIST[x]
    x=x+1
server_DF
```

```python
M=12
Inv_E=0
short=0
server_DF["Inv. B."][0]=0
server_DF["Inv. B."][2]=8
server_DF["Lead"][0]=1
```

```python
for i in range(0,len(server_DF)):
    
    Demand=server_DF["Demand"][i]
    Inv_B=server_DF["Inv. B."][i]
    
    if((Demand+short)<=Inv_B):
        server_DF["Inv.E."][i]=Inv_B-Demand-short
        Inv_E=server_DF["Inv.E."][i]
        try:    
            server_DF["Inv. B."][i+1]+=Inv_E
        except IndexError:
            print("IndexError: out of range")
        except KeyError:
            print("KeyError: out of range")
        Inv_B=Inv_E
        server_DF["Short"][i]=0
        short=0
    else:
        server_DF["Inv.E."][i]=0
        server_DF["Short"][i]=np.abs(Inv_B-Demand)+short
        short=server_DF["Short"][i]
        Inv_B=0
        Inv_E=0
    if((i+1 ) % 5 == 0):
        server_DF["Q_Order"][i]=M-Inv_E+short
        Q_o=server_DF["Q_Order"][i]
        Lead_conditions = [(server_DF["RN L"][i] >= low) & (server_DF["RN L"][i]<= high) for low, high in zip(Lead_time_df["ran"], Lead_time_df["range"])]

        server_DF["Lead"][i]=np.select(Lead_conditions, Lead_time_df["Lead time"])
        L_T=server_DF["Lead"][i]
        try:
            server_DF["Inv. B."][i+1+L_T]=server_DF["Inv. B."][i+1+L_T]+Q_o
        except IndexError:
            print("IndexError: out of range")
        except KeyError:
            print("KeyError: out of range")
    server_DF["Lead"][i+1]=server_DF["Lead"][i]-1
        
```

```python
AVG_ending_time=np.sum(server_DF["Inv.E."])/25
number_of_short_time=(server_DF['Short'] != 0).sum()
```

```python
server_DF['Lead'] = server_DF['Lead'].apply(lambda x: '-' if x < 0 else x)
server_DF["RN L"].replace(-1, "-", inplace=True)
```

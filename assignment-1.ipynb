```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python

IAT_df = pd.DataFrame.from_dict({'IAT': [1, 2, 3, 4], 'P': [0.25, 0.40, 0.20, 0.15]})
IAT_df['CDF'] = IAT_df['P'].cumsum()
IAT_df['range'] = IAT_df['CDF'] * 100  -1
IAT_df['ran']=[0] + IAT_df["CDF"].tolist()[:-1]

print(IAT_df.info())
IAT_df
```

```python
ABLE_ST_df = pd.DataFrame.from_dict({'IAT': [2, 3, 4,5], 'P': [0.3,0.28,0.25,0.17]})
ABLE_ST_df['CDF'] = ABLE_ST_df['P'].cumsum()
ABLE_ST_df['range'] = ABLE_ST_df['CDF'] * 100  -1
ABLE_ST_df['ran']=[0] + ABLE_ST_df["CDF"].tolist()[:-1]
print(ABLE_ST_df.info())
ABLE_ST_df
```

```python
BAKER_ST_df = pd.DataFrame.from_dict({'IAT': [2,3, 4,5], 'P': [0.35,0.25,0.20,0.20]})
BAKER_ST_df['CDF'] = BAKER_ST_df['P'].cumsum()
BAKER_ST_df['range'] = BAKER_ST_df['CDF'] * 100  -1
BAKER_ST_df['ran']=[0] + BAKER_ST_df["CDF"].tolist()[:-1]
print(BAKER_ST_df.info())
BAKER_ST_df
```

```python
IAT_RN_LIST=[0,26,98,90,26,42,74,80,68,22,48,34,45,24,34,63,38,80,42,56,89,18,51,71,16,92,0]
ST_RN_LIST= [95,21,51,92,89,38,13,61,50,49,39,53,88,1,81,53,81,64,1,67,1,47,75,57,87,47,0]
```

```python
double_server_DF=pd.DataFrame.from_dict({'RN_IAT':IAT_RN_LIST ,"IAT":None,"clock":None, 'RN_ST': ST_RN_LIST,"Able_ST_Begine":0,"Able_ST":0,"Able_ST_End":0,"Baker_ST_Begine":0,"Baker_ST":0,"Baker_ST_End":0,"Queuing_Time":0,"Time_Spent_n_System":0,"Able_IDT":-1,"Baker_IDT":-1})
double_server_DF.index.name = 'Call_ID'
double_server_DF
```

```python

IAT_conditions = [(double_server_DF["RN_IAT"] >= low) & (double_server_DF["RN_IAT"] <= high) for low, high in zip(IAT_df["ran"], IAT_df["range"])]

double_server_DF["IAT"] = np.select(IAT_conditions, IAT_df["IAT"])
```

```python
# Assuming the first row has a predefined initial value for "clock"
double_server_DF["clock"][0] = 0  # Replace 'initial_value' with the actual initial value

# Iterate over the rows starting from the second row
for i in range(1, len(double_server_DF)):
    double_server_DF["clock"][i] = double_server_DF["clock"][i - 1] + double_server_DF["IAT"][i].astype(float)
```

```python
Able_ST_End=0
Able_ST_Begine=0
clock=0
Baker_ST_End=0
Baker_ST_Begine=0
print("c = ",clock," A = ",Able_ST_End," B= ",Baker_ST_End)
for i in range(0,len(double_server_DF)-1):
    if ( ((clock-Able_ST_End) ==(clock-Baker_ST_End))or(((clock-Baker_ST_End))>=((clock-Able_ST_End))) or(((clock-Baker_ST_End))>=((0)))or(Baker_ST_End==clock)):
        double_server_DF["Baker_ST_Begine"][i] =np.max(np.array([Baker_ST_End, double_server_DF["clock"][i]]))
        Baker_conditions = [(double_server_DF["RN_ST"][i] >= low) & (double_server_DF["RN_ST"][i] <= high) for low, high in zip(BAKER_ST_df["ran"], BAKER_ST_df["range"])]
        double_server_DF["Baker_ST"][i] = np.select(Baker_conditions, BAKER_ST_df["IAT"])
        double_server_DF["Baker_ST_End"][i]=double_server_DF["Baker_ST_Begine"][i]+double_server_DF["Baker_ST"][i]
        double_server_DF["Baker_IDT"][i]=double_server_DF["Baker_ST_Begine"][i]-Baker_ST_End
        Baker_ST_End=double_server_DF["Baker_ST_End"][i]
        Baker_ST_Begine=double_server_DF["Baker_ST_Begine"][i]
        clock=double_server_DF["clock"][i+1]
        double_server_DF["Queuing_Time"][i]=np.abs(double_server_DF["Baker_ST_Begine"][i]-double_server_DF["clock"][i])

    else:
        double_server_DF["Able_ST_Begine"][i] = np.max(np.array([Able_ST_End, double_server_DF["clock"][i]]))
        Able_conditions = [(double_server_DF["RN_ST"][i] >= low) & (double_server_DF["RN_ST"][i] <= high) for low, high in zip(ABLE_ST_df["ran"], ABLE_ST_df["range"])]
        double_server_DF["Able_ST"][i] = np.select(Able_conditions, ABLE_ST_df["IAT"])
        double_server_DF["Able_ST_End"][i]=double_server_DF["Able_ST_Begine"][i]+double_server_DF["Able_ST"][i]
        double_server_DF["Able_IDT"][i]=double_server_DF["Able_ST_Begine"][i]-Able_ST_End
        Able_ST_End=double_server_DF["Able_ST_End"][i]
        Able_ST_Begine=double_server_DF["Able_ST_Begine"][i]
        clock=double_server_DF["clock"][i+1]
        double_server_DF["Queuing_Time"][i]=np.abs(double_server_DF["Able_ST_Begine"][i]-double_server_DF["clock"][i])
    print("c = ",clock," A = ",Able_ST_End," B= ",Baker_ST_End)
```

```python
double_server_DF["Time_Spent_n_System"] = np.maximum(double_server_DF["Baker_ST"], double_server_DF["Able_ST"]) + double_server_DF["Queuing_Time"]
```

```python
double_server_DF["RN_IAT"].replace(-1, "-", inplace=True)

double_server_DF["Able_ST_Begine"].replace(0, "-", inplace=True)
double_server_DF["Able_ST"].replace(0, "-", inplace=True)
double_server_DF["Able_ST_End"].replace(0, "-", inplace=True)

double_server_DF["Baker_ST_Begine"].replace(0, "-", inplace=True)
double_server_DF["Baker_ST"].replace(0, "-", inplace=True)
double_server_DF["Baker_ST_End"].replace(0, "-", inplace=True)

double_server_DF["Able_IDT"].replace(-1, "-", inplace=True)
double_server_DF["Baker_IDT"].replace(-1, "-", inplace=True)

double_server_DF["Baker_ST_Begine"][0]=0
```
